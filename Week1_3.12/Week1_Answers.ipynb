## Exercise 4

Make a new dataframe of only rows that have a last_review in 2019 or 2020.
Hint, you could convert the last listings date column to have date time objects. 
https://www.journaldev.com/23365/python-string-to-datetime-strptime

Or you could use string operations, string.split would be helpful.


```python
# using datetime, a good library to know

# my first attempt gives an error because strptime needs two arguments, the solution is a lambda function
# df['last_review_datetime'] = df.last_review.astype(str).apply(datetime.strptime('%Y-%m-%d'))

# my second attempt also gives an illustrative error, I didn't account for the nans in the column!
# df['last_review_datetime'] = df.last_review.apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d'))

df['last_review_datetime'] = df.last_review.apply(lambda x: np.nan if pd.isna(x) else datetime.strptime(str(x), '%Y-%m-%d'))
# Note that for **reasons** datetime columns have NaT instead of NaN, they can be expected to work the same

df[df.last_review_datetime >= datetime(2019, 1, 1)]
```

```python
# keeping everything a string
df['last_review_year'] = df.last_review.apply(lambda x: np.nan if pd.isna(x) else str(x).split('-')[0])

df[(df.last_review_year == '2019') | (df.last_review_year == '2020')]
```
